```python
from scipy.io import wavfile
import pandas as pd
samplerate, data = wavfile.read('./vtech_preds_late.wav')
```

```python
target = data[3260000:3285925, 0]
p_clutch = data[3260000:3285925, 1]
b_clutch = data[3260000:3285925, 2]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.plot(target)
#plt.xlabel("Time (samples)")
plt.ylabel("Clutch State")
plt.title("Desired Clutch Behavior")

plt.subplot(3,1,2)
plt.plot(p_clutch, "C1")
#plt.xlabel("Time (samples)")
plt.ylabel("Probability")
plt.title("Predicted Clutch State")
plt.grid()

plt.subplot(3,1,3)
plt.plot(b_clutch, "C2")
plt.xlabel("Time (samples)")
plt.ylabel("Clutch State")
plt.title("Actual Clutch Behavior")

plt.tight_layout()
```
